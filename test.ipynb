```python
import requests

url = "https://www.bing.com/news/search?q=earthquake+delhi"

response = requests.get(url)
data = response.text

# Process the scraped data here
# ...
data
```

```python

from bs4 import BeautifulSoup
soup = BeautifulSoup(data, 'html.parser')
caption_divs = soup.find_all('div', class_='caption')

# Extract the HTML content from each caption div
html_content_list = [div.prettify() for div in caption_divs]

# Print the HTML content
for html_content in html_content_list:
    print(html_content)
```

```python
# Assuming you have the HTML content stored in the variable 'data'
soup = BeautifulSoup(data, 'html.parser')

# Find all <span> tags with area-label attribute
span_tags = soup.find_all('span', {'tabindex': True})

span_tags
# Extract the text from each <span> tag
span_text_list = [tag.text for tag in span_tags]
span_text_list = [text for text in span_text_list if len(text) > 1]

# Print the combined extracted text
print(span_text_list)
```

```python
from bs4 import BeautifulSoup

# Assuming you have the HTML content stored in the variable 'data'
soup = BeautifulSoup(data, 'html.parser')

# Find all <a> tags
a_tags = soup.find_all('a')

# Extract the text from each <a> tag
text_list = [tag.text for tag in a_tags]
```

```python
# Assuming you have the HTML content stored in the variable 'data'
soup = BeautifulSoup(data, 'html.parser')

# Find all <a> tags
a_tags = soup.find_all('a')

# Extract the text from each <a> tag
text_list = []
for tag in a_tags:
    span_tag = tag.find('span')
    if span_tag:
        text_list.append(span_tag.text)

text_list = [text for text in text_list if len(text.split()) > 2]
text_list
len(text_list)
```
